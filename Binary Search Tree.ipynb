```python
class Node:
    
    def __init__(self,data):
        self.data = data
        self.left = None
        self.right = None
          
def take_input():
    
    data = int(input())
    
    if data == -1:
        return None
    
    root = Node(data)
    
    Left = take_input()
    Right = take_input()
    
    root.left = Left
    root.right = Right
    
    return root
    
    
def printTree(root):
    
    if root == None:
        return 
    
    print(root.data, end=" ")
    
    if root.left != None:
        print("L", root.left.data, end=":")
        
    if root.right != None:
        print("R", root.right.data, end=" ")
    
    print()
    printTree(root.left)
    printTree(root.right)    
        
def Search_in_BST(root,target):
    
    if root == None:
        return False

    if root.data == target:
        return True
    
    elif root.data < target:
        Search_in_BST(root.right,target)
        
    else:
        Search_in_BST(root.left,target)
```

```python
class Node:
    
    def __init__(self,data):
        self.data = data
        self.left = None
        self.right = None

def buildTree(arr):
    
    if not arr:
        return None
    
    mid = len(arr)//2
    
    root = Node(arr[mid])
    root.left = buildTree(arr[:mid])
    root.right = buildTree(arr[mid+1:])
    
    return root

def preOrder(root):
    if root == None:
        return 
    print(root.data,end=" ")
    preOrder(root.left)
    preOrder(root.right)
```

```python
arr = [1,2,3,4,5,6,7]
root = buildTree(arr)
preOrder(root)
```

```python
def minInTree(root):
    
    if root == None:
        return 10000000

    leftTree = minInTree(root.left)
    rightTree = minInTree(root.right)
    return min(leftTree,rightTree,root.data)


def maxInTree(root):
    if root == None:
        return -100000000
    leftTree = maxInTree(root.left)
    rightTree = maxInTree(root.right)
    return max(leftTree,rightTree,root.data)
    

def isBST(root):
    
    if root == None:
        return True
    
    leftMax = maxInTree(root.left)
    rightMin = minInTree(root.right)
    
    if root.data <= leftMax or root.data > rightMax:
        return False
    
    isLeftBST = isBST(root.left)
    isRightBST = isBST(root.right)
    
    return isLeftBST and isRightBST
```

```python
def returnPath(root,target):
    
    if root == None:
        return None
    
    if root.data == target:
        l = list()
        l.append(root.data)
        return l   
    
    leftOutput = returnPath(root.left,target)
    if leftOutput != None:
        leftOutput.append(root.data)
        return leftOutput
    
    rightOutput = returnPath(root.right,target)
    if rightOutput != None:
        rightOutput.append(root.data)
        return rightOutput    
```
