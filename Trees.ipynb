```python
 class Node:
        
        def __init__(self,data):
            self.data = data
            self.left = None
            self.right = None
            
def printTree(root):
    
    if root == None:
        return 
    print(root.data, end=":")
    if root.left != None:
        print("L",root.left.data, end=":")
    if root.right != None:
        print("R",root.right.data, end=":")
    print()
    printTree(root.left)
    printTree(root.right)    
    
def takeInput():
    
    rootData = int(input())
    if rootData == -1:
        return None
    
    root = Node(rootData)
    leftTree = takeInput()
    rightTree = takeInput()
    
    root.left = leftTree
    root.right = rightTree
    
    return root

## Count the number of node
def numNode(root):
    
    if root == None:
        return 0
    
    leftCount = numNode(root.left)
    rightCount = numNode(root.right)
    return 1 + leftCount + rightCount

#Pre-Order -----------------------------

def preOrder(root):
    
    if root == None:
        return
    print(root.data)
    preOrder(root.left)
    preOrder(root.right)
    return

#Post-Order ---------------------------------

def postOrder(root):
    
    if root == None:
        return 
    postOrder(root.left)
    postOrder(root.right)
    print(root.data)
    return

#In-Order -----------------------------------

def inOrder(root):
    
    if root == None:
        return 
    
    inOrder(root.left)
    print(root.data)
    inOrder(root.right)
    return

def largestData(root):
    
    
    if root == None:
        return -1
    
    leftLarge = largestData(root.left)
    rightLarge = largestData(root.right)
    largest = max(root.data,leftLarge,rightLarge)
    return largest
    
## Height of a Tree:
    
def height(root):
    
    if root == None:
        return 0
    
    L = height(root.left)
    R = height(root.right)
    return 1 + max(L,R)

## Count Leaf Node:

def leafNodesCount(root):
    
    if root == None:
        return 0
    
    if root.left == None and root.right == None:
        return 1
  
    left = leafNodesCount(root.left)
    right = leafNodesCount(root.right)

    return left + right
```

```python
def printAtK(root,k,d=0):
    
    if root == None:
        return
    
    if k == d:
        print(root.data)
        return
    left = printAtK(root.left,k,d+1)
    right = printAtK(root.right,k,d+1)
```
